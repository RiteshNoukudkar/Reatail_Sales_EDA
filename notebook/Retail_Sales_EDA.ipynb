## Month and Quarter Conversion

```python
df['Month'] = df['Date'].dt.to_period('M')
df['Quarter'] = df['Date'].dt.to_period('Q')
```

## Monthly Sales

```python
monthly_sales = df.groupby('Month')['Total Amount'].sum()
monthly_sales
```

## Observation

- Sales changed from month to month throughout the year.
- The highest sales were recorded in May.
- Some months had lower sales than others, showing seasonal fluctuations.
- These trends can help the company plan future marketing strategies and inventory management.

```python
monthly_sales = df.groupby("Month")["Total Amount"].sum()

monthly_sales
```

## Monthly Sales Line Chart

```python
plt.figure(figsize=(10,5))

plt.plot(
    monthly_sales.index.astype(str),
    monthly_sales.values,
    marker='o',
    linewidth=2
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

plt.grid(True)

plt.show()
```

## Observation

- Monthly sales show fluctuations across different months.
- Some months generated higher revenue compared to others.
- The sales trend indicates changes in customer purchasing behaviour over time.
- Peak sales months can be targeted for inventory expansion and marketing campaigns.
- Low-performing months may require promotional offers to increase sales.

# Quarterly Sales Trend

```python
# Create Quarter column
df["Quarter"] = df["Date"].dt.to_period("Q")

df.head()
```

## Calculate Quarterly Sales

```python
# Calculate Quarterly Sales
quarterly_sales = df.groupby("Quarter")["Total Amount"].sum()

quarterly_sales
```

## Quarterly Sales Trends Line Chart 

```python
plt.figure(figsize=(8,5))

plt.plot(
    quarterly_sales.index.astype(str),
    quarterly_sales.values,
    marker='o',
    linewidth=2
)
plt.xlabel("Quarter")
plt.ylabel("Total Sales")
plt.title("Quarterly Sales Trend")  

plt.grid(True)

plt.show()
```

## Observation

- The quarterly sales trend provides an overview of business performance over each quarter.
- Sales vary across different quarters, indicating seasonal customer demand.
- Higher sales quarters may correspond to festive seasons, marketing campaigns, or increased customer activity.
- Lower-performing quarters may require promotional strategies to improve revenue.

## Find the Top Product Category

```python
top_products = (
    df.groupby("Product Category")["Quantity"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,5))
sns.barplot(x=top_products.values, y=top_products.index)

plt.title("Top 10 Best-Selling Product Categories")
plt.xlabel("Total Quantity Sold")
plt.ylabel("Product Category")

plt.show()
```

## Observation

- Some product categories consistently outperform others in terms of sales volume.
- High-selling categories contribute significantly to overall business performance.
- The company should maintain inventory levels and marketing efforts for these top-performing categories.

## Corelation Hitmap

```python
correlation = df.select_dtypes(include=['number']).corr()
correlation
```

## heatmap 

```python
plt.Figure(figsize=(10,6))

sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm',   
    fmt='.2f'
)

plt.title("Correlation Heatmap Matrix")
plt.show()
```

## Observation

- The heatmap shows the relationship between numerical variables.
- Positive correlation indicates that two variables increase together.
- Negative correlation indicates that one variable increases while the other decreases.
- Strong correlations help identify the most influential factors affecting sales.
- Weak correlations suggest little or no linear relationship between variables.

## Average Purchase by Gender

```python
plt.figure(figsize=(6,5))

sns.barplot(
    data=df,
    x="Gender",
    y="Total Amount",
    estimator="mean"
)

plt.title("Average Purchase Amount by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Purchase Amount")

plt.show()
```
